```python
features = ['price', 'availability_365', 'number_of_reviews_ltm', 'minimum_nights','number_of_reviews']

def analyze_distribution(data, feature):
    plt.figure(figsize=(18, 6))
    # Boxplot
    plt.subplot(1, 3, 2)
    sns.boxplot(x=df[feature].dropna(), color='orange')
    plt.title(f"Boxplot of {feature}")
    plt.xlabel(feature)

    # QQ Plot
    plt.subplot(1, 3, 3)
    probplot(df[feature].dropna(), dist="norm", plot=plt)
    plt.title(f"QQ Plot of {feature}")

    plt.tight_layout()
    #plt.savefig('distribusi.png', dpi=300, bbox_inches='tight')
    plt.show()
    

for feature in features:
    analyze_distribution(df, feature)
```

<blockquote style="background: #e9edc9; padding: 10px; border-left: 4px solid #ccd5ae; margin: 10px 0;">
    <b>INSIGHT</b>
    <li>Distribusi Data Tidak Normal: Hampir semua variabel, termasuk `price`,`availability_365` ,`number_of_reviews_ltm`,`minimum_nights`, dan `number_of_reviews`, menunjukkan distribusi yang tidak normal dengan banyak outlier, terutama pada kuantil tinggi.</li>
    <li>Keberadaan Outlier Signifikan: Variabel seperti `price`, `minimum_nights`, dan `number_of_reviews` memiliki outlier yang sangat besar, yang dapat mendistorsi hasil analisis statistik jika tidak ditangani dengan baik.</li>
    <li>Sebagian besar data menunjukkan tren rendah, seperti harga yang rendah, jumlah ulasan sedikit, dan waktu minimum inap pendek.
    Hanya sebagian kecil listing yang memiliki nilai ekstrem, seperti harga tinggi atau jumlah ulasan besar.</li>

```python
def analyze_distribution(data, feature):
    print(f"\n=== Analisis Distribusi untuk Fitur: {feature} ===")
    stat, p_value = shapiro(data[feature].dropna())
    print(f"Shapiro-Wilk Test: Statistik={stat:.4f}, p-value={p_value:.4f}")

    if p_value < 0.05:
        print(f"Hasil: Data {feature} **tidak berdistribusi normal** (p < 0.05)")
    else:
        print(f"Hasil: Data {feature} **berdistribusi normal** (p >= 0.05)")

for feature in features:
    analyze_distribution(df, feature)
```

```python
# mengidentifikasi missing values berdasarkan jumlah dan proporsi
def missing_values_table(data):
    miss_val_count = data.isnull().sum()
    miss_val_percent = ((miss_val_count / len(data)) * 100).round(2).astype(str)+'%'
    table = pd.concat([miss_val_count, miss_val_percent], axis=1)
    table.columns = ['Jumlah Missing Values', 'Proporsi Missing Values']
    table = table[table['Jumlah Missing Values'] != 0].sort_values('Proporsi Missing Values', ascending=False).round(2)
    return table

df_missing_values = missing_values_table(df)
display(df_missing_values)
```

> Berdasarkan informasi [source](https://www.mdpi.com/2072-4292/12/22/3802), District yang ada di Kota Bangkok dapat dibagi menjadi 3 bagian yaitu inner zone, middle zone dan outer zone. Dengan memberikan kategori tersebut akan mempermudah dalam melakukan analisis nantinya.
</p>

```python
def label(neighbourhood):
    if neighbourhood in ['Bang Phlat','Chatu Chak','Bang Sue','Dusit','Bang Rak','Bangkok Noi',
                                'Bangkok Yai','Khlong Toei','Ratchathewi','Parthum Wan','Pom Prap Sattru Phai','Phra Nakhon','Phaya Thai','Huai Khwang','Din Daeng','Sathon','Khlong San','Thon buri','Bang Kho laen','Yan na wa','Samphanthawong','Vadhana']:
        return 'Inner'
    elif neighbourhood in ['Taling Chan','Thawi Watthana','Nong Khaem','Lak Si','Don Mueang',
                                  'Sai Mai','Bang Kapi','Lat Phrao','Bang Khen','Bueng Kum','Chom Thong',
                                  'Bang Na','Pra Wet','Suanluang','Saphan Sung','Khan Na Yao','Wang Thong Lang',
                                  'Thung khru','Bang Khae','Phra Khanong','Rat Burana','Phasi Charoen']:
        return 'Urban Fringe'
    elif neighbourhood in ['Nong Chok','Lat Krabang','Min Buri','Khlong Sam Wa','Bang Khun thain',
                                  'Bang Bon']:
        return 'Suburban'
    else:
        return 'Unknown'

df['district'] = df['neighbourhood'].apply(label)
```

```python
from collections import Counter
import re

# Gabungkan semua teks dari kolom 'name' menjadi satu string
all_text = ' '.join(df['name'].astype(str))

# Preprocessing teks: konversi menjadi huruf kecil dan hapus karakter khusus
clean_text = re.sub(r'[^a-zA-Z\s]', '', all_text.lower())

# Tokenisasi: pisahkan teks menjadi kata-kata
tokens = clean_text.split()

# Hitung frekuensi masing-masing kata
word_freq = Counter(tokens)

# Identifikasi kata paling sering digunakan
most_common_words = word_freq.most_common(100)  #

# Kata kunci
keywords_aksesibilitas = ['bts', 'mrt', 'near', 'min', 'walk', 'mins', 'station', 'airport', 'close', 'skytrain', 'city center', 'downtown', 'heart of', 'prime location']
keywords_promosi = ["deal", "promo"]
keywords_fasilitas = ['pool', 'wifi', 'gym', 'rooftop', 'park', 'king bed', 'queen bed', 'sofa bed', 'private bathroom', 'family room', 'pet friendly']
keywords_estetika = ['new', 'cozy', 'luxury', 'view', 'spacious', 'modern', 'big', 'clean', 'lovely', 'comfy', 'nice', 'quiet', 'peaceful', 'green', 'garden view', 'breakfast', 'park view']
keywords_pengalaman = ['local', 'authentic', 'cultural', 'traditional', 'unique', 'heritage', 'home-stay']
keywords_keamanan = ['secure', 'safe', 'privacy', 'private', 'key card', '24/7 security']
keywords_premium = ['penthouse', 'suite', 'exclusive', 'premium', 'high floor', 'skyline']

# Fungsi untuk mencocokkan kata kunci secara eksak (case-insensitive)
def check_exact_keywords(name, keywords):
    words = re.findall(r'\b\w+\b', name.lower())  
    return any(word in keywords for word in words)
df['keyword'] = 'tidak memakai keywords'

# Klasifikasikan berdasarkan kategori kata kunci
for index, row in df.iterrows():
    description = row['name'].lower()  
    if check_exact_keywords(description, keywords_aksesibilitas):
        df.at[index, 'keyword'] = 'aksesibilitas'
    elif check_exact_keywords(description, keywords_promosi):
        df.at[index, 'keyword'] = 'promosi'
    elif check_exact_keywords(description, keywords_fasilitas):
        df.at[index, 'keyword'] = 'fasilitas'
    elif check_exact_keywords(description, keywords_estetika):
        df.at[index, 'keyword'] = 'estetika'
    elif check_exact_keywords(description, keywords_pengalaman):
        df.at[index, 'keyword'] = 'pengalaman'
    elif check_exact_keywords(description, keywords_keamanan):
        df.at[index, 'keyword'] = 'keamanan'
    elif check_exact_keywords(description, keywords_premium):
        df.at[index, 'keyword'] = 'premium'

df.head(5)
```
